```python
from bs4 import BeautifulSoup as bs
import requests

url = "https://movie.naver.com/movie/bi/mi/pointWriteFormList.nhn?code=189000&type=after&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false"
html = bs(requests.get(url).content, "html.parser")
cnt = html.select('body > div > div > div.score_total > strong > em')[0].contents[0].replace(',', '')

review = []

for x in range(1, int(cnt)//10+1):
    html = bs(requests.get(url + "&page=" + str(x)).content, "html.parser")
    reple = html.select(".score_result .score_reple p")
    
    for i in reple:
        result = i.get_text()
        result = result.replace('\t','')
        result = result.replace('\r','')
        result = result.replace('\n','')
        review.append(result)

f = open('data1.txt', 'w', encoding='utf-8')

for r in review:
    f.write(r)
    
f.close()
```

```python
from konlpy.tag import Okt 
from collections import Counter

stopwords = ['내','그','말','부분','엔딩','대한','이건','대해','보지','시작','상미','살짝','결말','스포','영화로','이영화','끝','더','좀','끝','전혀','더','마지막','다시','나름','만','왜','약간','제대로','매우','꼭','초반','스토리','보기','처음','중간','내용','이야기','중반','조금','모두','모든','계속','가장','최근','제일','후반','내내','지금','역시','때문','영화', '관람객', '감상', '수', '때', '이', '것', '포함', '평점', '보고', '진짜', '스포일러','평','정말','아주','볼']
okt = Okt()
noun = okt.nouns(data)
noun = [n for n in noun if n not in stopwords]

for i,v in enumerate(noun):
    if len(v) < 2:
        noun.pop(i)
        
count = Counter(noun)
```

```python
from bs4 import BeautifulSoup as bs
import requests

url = "https://movie.naver.com/movie/bi/mi/pointWriteFormList.nhn?code=97612&type=after&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false"
html = bs(requests.get(url).content, "html.parser")
cnt = html.select('body > div > div > div.score_total > strong > em')[0].contents[0].replace(',', '')

review = []

for x in range(1, int(cnt)//10+1):
    html = bs(requests.get(url + "&page=" + str(x)).content, "html.parser")
    reple = html.select(".score_result .score_reple p")
    
    for i in reple:
        result = i.get_text()
        result = result.replace('\t','')
        result = result.replace('\r','')
        result = result.replace('\n','')
        review.append(result)

f = open('data10.txt', 'w', encoding='utf-8')

for r in review:
    f.write(r)
    
f.close()
```

```python
f = open('data10.txt', 'r', encoding='utf-8')
data = f.read()
```

```python
from konlpy.tag import Okt 
from collections import Counter

stopwords = ['더','대한','마치','이','모든','내내','영화','관람객','스토리','상미','정말','내용','이야기','보고','평점','미가','진짜','중간','처음','모두','다시','한편','역시','한번','수','보기','조금','아주','약간','것']
okt = Okt()
noun = okt.nouns(data)
noun = [n for n in noun if n not in stopwords]

for i,v in enumerate(noun):
    if len(v) < 2:
        noun.pop(i)
        
count = Counter(noun)
```

```python
noun_list = count.most_common(100)
for v in noun_list:
    print(v)
```

```python
import csv
with open('movie10_noun_list.csv', 'w', encoding='utf-8') as f:
    f.write('word,freq\n')
    writer = csv.writer(f)
    writer.writerows(noun_list)
```
